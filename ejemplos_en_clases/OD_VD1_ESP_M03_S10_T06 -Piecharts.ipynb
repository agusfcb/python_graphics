### Creación de Piecharts

Importación de Bibliotecas Necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de Datos de Ejemplo


```python
etiquetas = ['A', 'B', 'C', 'D']
tamaños = [15, 30, 45, 10]
```

Creación de un Gráfico de Pie Básico


```python
plt.pie(tamaños, labels=etiquetas)
plt.title('Gráfico de Pie Básico')
plt.show()
```

### Personalización de Gráficos de Pie
Añadir Etiquetas y Porcentajes

```python
plt.pie(tamaños, labels=etiquetas, autopct='%1.1f%%')
plt.title('Gráfico de Pie con Porcentajes')
plt.show()
```

Cambiar Colores y Estilos


```python
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140)
plt.title('Gráfico de Pie Personalizado')
plt.show()
```

Explosión de Sectores


```python
explosion = (0.1, 0, 0, 0)  # "explota" el 1er sector
plt.pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140, explode=explosion)
plt.title('Gráfico de Pie con Explosión de Sector')
plt.show()
```

Gráficos de Pie con Subplots


```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Gráfico de Pie 1')

axs[1].pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Gráfico de Pie 2')

plt.tight_layout()
plt.show()
```
