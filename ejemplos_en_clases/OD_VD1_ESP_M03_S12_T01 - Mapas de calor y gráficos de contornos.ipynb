### Mapas de calor y gráficos de contorno

Importamos las librerías necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

Mapa de calor

```python


# Generar datos de ejemplo
x = np.linspace(-3, 3, 100)   # Array de 100 puntos equiespaciados entre -3 y 3
y = np.linspace(-3, 3, 100)   # Array de 100 puntos equiespaciados entre -3 y 3
X, Y = np.meshgrid(x, y)      # Crear una malla de coordenadas X, Y

# Calcular valores Z para el mapa de calor y el gráfico de contorno
Z = np.sin(X) * np.cos(Y)     # Función de ejemplo para Z

# Crear figura y ejes para el mapa de calor
fig, ax_heatmap = plt.subplots(figsize=(6, 5))  # Figura con un solo subplot para el mapa de calor

# Dibujar mapa de calor
heatmap = ax_heatmap.imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap='RdBu')
# imshow() dibuja el mapa de calor con Z
# extent=[-3, 3, -3, 3] define los límites de los ejes X e Y
# origin='lower' coloca el origen del gráfico en la esquina inferior izquierda
# cmap='RdBu' define el esquema de color (rojo-blanco-azul)

# Añadir barra de color
plt.colorbar(heatmap, ax=ax_heatmap)  # Añade la barra de color asociada al mapa de calor
ax_heatmap.set_title('Mapa de Calor')  # Título del subplot
ax_heatmap.set_xlabel('Eje X')         # Etiqueta del eje X
ax_heatmap.set_ylabel('Eje Y')         # Etiqueta del eje Y
```

Gráfico de contorno

```python

# Crear figura y ejes para el gráfico de contorno
fig, ax_contour = plt.subplots(figsize=(6, 5))  # Figura con un solo subplot para el gráfico de contorno

# Dibujar gráfico de contorno
contour = ax_contour.contour(X, Y, Z, levels=10, cmap='RdBu')
# contour() dibuja el gráfico de contorno con X, Y y Z
# levels=10 define el número de niveles de contorno a mostrar
# cmap='RdBu' define el esquema de color (rojo-blanco-azul)

# Añadir barra de color
plt.colorbar(contour, ax=ax_contour)  # Añade la barra de color asociada al gráfico de contorno
ax_contour.set_title('Gráfico de Contorno')  # Título del subplot
ax_contour.set_xlabel('Eje X')               # Etiqueta del eje X
ax_contour.set_ylabel('Eje Y')               # Etiqueta del eje Y

plt.tight_layout()  # Ajusta automáticamente la disposición de los subplots para que no se superpongan
plt.show()         # Muestra la figura con los subplots
```
