### Creación de Boxplots y Violin plots

Importación de Bibliotecas Necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de Datos de Ejemplo


```python
np.random.seed(0)
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

Creación de un Boxplot Básico


```python
plt.boxplot(datos)
plt.title('Boxplot Básico')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

### Creación de Gráficos de Violín Básicos
Creación de un Gráfico de Violín Básico

```python
plt.violinplot(datos)
plt.title('Gráfico de Violín Básico')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

### Comparación de Distribuciones
Boxplots Múltiples para Comparación

```python
datos2 = [np.random.normal(1, std, 100) for std in range(1, 4)]

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].boxplot(datos)
axs[0].set_title('Boxplots Grupo 1')
axs[0].set_xlabel('Grupo')
axs[0].set_ylabel('Valores')

axs[1].boxplot(datos2)
axs[1].set_title('Boxplots Grupo 2')
axs[1].set_xlabel('Grupo')
axs[1].set_ylabel('Valores')

plt.tight_layout()
plt.show()
```

Gráficos de Violín Múltiples para Comparación


```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].violinplot(datos)
axs[0].set_title('Gráficos de Violín Grupo 1')
axs[0].set_xlabel('Grupo')
axs[0].set_ylabel('Valores')

axs[1].violinplot(datos2)
axs[1].set_title('Gráficos de Violín Grupo 2')
axs[1].set_xlabel('Grupo')
axs[1].set_ylabel('Valores')

plt.tight_layout()
plt.show()
```

### Personalización de Boxplots y Gráficos de Violín
Cambiar Colores y Estilos

```python
plt.boxplot(datos)
plt.title('Boxplot con Títulos y Etiquetas')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

Personalización de Ejes y Leyendas


```python
plt.boxplot(datos, patch_artist=True, labels=['A', 'B', 'C'],
            boxprops=dict(facecolor='lightgreen', color='green'),
            whiskerprops=dict(color='green'),
            capprops=dict(color='green'),
            medianprops=dict(color='red'))
plt.title('Boxplot con Leyenda y Etiquetas')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

### Técnicas Avanzadas en Boxplots y Gráficos de Violín
Boxplots con Datos Agrupados

```python
datos_agrupados = [np.random.normal(loc, scale, 100) for loc, scale in zip(range(1, 4), range(1, 4))]
plt.boxplot(datos_agrupados, patch_artist=True, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'],
            boxprops=dict(facecolor='lightcoral', color='darkred'),
            whiskerprops=dict(color='darkred'),
            capprops=dict(color='darkred'),
            medianprops=dict(color='yellow'))
plt.title('Boxplot con Datos Agrupados')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

Gráficos de Violín con Datos Agrupados


```python
plt.violinplot(datos_agrupados, showmeans=True)
plt.title('Gráfico de Violín con Datos Agrupados')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

Subplots con Boxplots y Gráficos de Violín


```python
#Recordatorio, los datos fueron generados anteriormente
fig, axs = plt.subplots(2, 1, figsize=(8, 12))

axs[0].boxplot(datos, patch_artist=True, labels=['A','B','C'],
               boxprops= dict(facecolor='lightblue', color='blue'),
               whiskerprops=dict(color='blue'),
               capprops=dict(color='blue'),
               medianprops=dict(color='red'))

axs[0].set_title('Boxplot')

axs[1].violinplot(datos, showmeans=True)
axs[1].set_title('Gráfico de Violín')

plt.tight_layout() # ubica los graficos para que no se solapen
plt.show()
```

### Ejemplos Prácticos
Ejemplo Básico

```python
plt.boxplot(datos)
plt.title('Ejemplo Básico de Boxplot')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

Ejemplo Avanzado con Personalización


```python
plt.violinplot(datos, showmeans=True, showmedians=True)
plt.title('Ejemplo Avanzado de Gráfico de Violín')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```
