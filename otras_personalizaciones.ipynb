```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 1, 2, 3, 4]

plt.plot(x, y1, label='Cuadrado')
plt.plot(x, y2, label='Línea')

plt.legend()
plt.title('Gráfico con Leyenda Básica')
plt.show()
```

```python
#Para modificar la ubicacion de la leyenda se usa el parametro loc
plt.plot(x, y1, label='Cuadrado')
plt.plot(x, y2, label='Línea')

plt.legend(loc='upper left')
plt.title('Gráfico con Leyenda en la Esquina Superior Izquierda')

plt.show()

plt.plot(x,y1, label='linea')
plt.legend(loc='center left')
plt.title('Gráfico con Leyenda en el Centro Izquierdo')
plt.show()
```

Las opciones de posición incluyen:

* 'best' (automático)
* 'upper right'
* 'upper left'
* 'lower left'
* 'lower right'
* 'right'
* 'center left'
* 'center right'
* 'lower center'
* 'upper center'
* 'center'

```python
#Tambien se puede personalizar las leyendas en sus propiedades como
# tamaño, fuente, titulo, otros
plt.plot(x, y1, label='Cuadrado', color='blue')
plt.plot(x, y2, label='Línea', color='green')

plt.legend(loc='best', fontsize='large', title='Tipos de Gráficos', title_fontsize='medium')
plt.title('Gráfico con Leyenda Personalizada')
plt.show()
```

```python
#Algo ya visto que son las anotaciones en los gráficos
plt.plot(x, y1, label='Cuadrado')
plt.plot(x, y2, label='Línea')

# el parametro xy es la ubicacion del punto
# el parametro xytext es la ubicacion del texto
# el parametro arrowprops es un diccionario con las caracteristicas del apuntador o flecha
plt.annotate('Punto destacado', xy=(3, 9), xytext=(3, 12), arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Gráfico con Anotación Básica')
plt.show()
```

```python
#Mayor personalizacion de la flecha marcadora
plt.plot(x, y1, label='Cuadrado')
plt.plot(x, y2, label='Línea')

plt.annotate('Punto destacado', xy=(3, 9), xytext=(2, 14),
             arrowprops=dict(facecolor='red', shrink=0.05, width=2),
             fontsize=12, color='black', backgroundcolor='yellow')
plt.title('Gráfico con Anotación Personalizada')
plt.show()
```

```python
plt.plot(x, y1, label='Cuadrado')
plt.plot(x, y2, label='Línea')

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y1[i]})', xy=(x[i], y1[i]), textcoords='offset points')

plt.title('Gráfico con Múltiples Anotaciones')
plt.show()
```

```python
#Ejemplo avanzado

y3 = [(d*2) for d in y2]

plt.plot(x, y1, label='Cuadrado', linestyle='-.', color='purple', marker='o')

plt.plot(x, y2, label='Línea', linestyle='--', color='green', marker='s')

plt.plot(x, y3, label='Linea', linestyle=':', color='red', marker='^')

plt.legend(loc='best', fontsize='large', title='Tipos de Gráficos', title_fontsize='medium')

plt.annotate('Punto Máximo', xy=(3, 9), xytext=(2, 14),
             arrowprops=dict(facecolor='red', shrink=0.05, width=2),
             fontsize=12, color='red', backgroundcolor='yellow')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.title('Ejemplo Avanzado con Leyendas y Anotaciones Personalizadas')
plt.show()
```
