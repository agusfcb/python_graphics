```python
#Los pairplots se utilizan para encontrar
# rapidamente relaciones entre variables.
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'iris' de Seaborn
iris = sns.load_dataset('iris')

iris
```

```python
# Crear un pairplot
sns.pairplot(data=iris, hue='species', diag_kind='hist', markers=["o", "s", "D"])

# Mostrar el grÃ¡fico
plt.show()
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='hist')
```

```python
#Se puede modificar el grafico de la diagonal con diag_kind
sns.pairplot(data=iris, hue='species', diag_kind='kde')
```

```python
# Si bien por defecto los diagramas suelen ser scatter
#   existen otras opciones
sns.pairplot(data=iris, hue='species', kind='scatter')
```

```python
# Aqui por ejemplo son diagramas con regresion lineal
sns.pairplot(data=iris, hue='species', kind='reg')
```
