```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el dataset 'tips' de Seaborn
tips = sns.load_dataset('tips')

# Crear un regplot de la propina en funcin del total de la cuenta
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips)

# Aadir ttulo y etiquetas
plt.title('Relacin entre Total de la Cuenta y Propina')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

# Mostrar el grfico
plt.show()
```

```python
# Crear un regplot avanzado con ajuste polinmico

plt.figure(figsize=(8, 6))

sns.regplot(x='total_bill', y='tip', data=tips, order=2, ci=None, scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Aadir ttulo y etiquetas
plt.title('Relacin entre Total de la Cuenta y Propina (Regresin Polinmica)')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

# Mostrar el grfico
plt.show()
```

```python
# Cargamos el dataset penguins
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
plt.title("Grafico inicial de lmplot")
plt.show()
```

```python
# Tambien permite agrupar con hue=

sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title("Regresion por especies")
plt.show()
```

```python
# Otro uso muy util que tiene lmplot y no regplot
#   es la posibilidad de agrupar y mostrar en varios graficos

sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
    hue='species', col='sex', height=4)

plt.show()
```

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
    col='species', row='sex', height=3)

plt.show()

#Para que los graficos tengan escalas independientes
# se debe agregar el parametro facet_kws
sns.lmplot(data=penguins, x='bill_length_mm', y='bill_depth_mm',
    col='species', row='sex', height=3, facet_kws=dict(sharex=False, sharey=False))

plt.show()
```

```python
# Crear un lmplot de la propina en funcin del total de la cuenta, separado por sexo
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips, height=6, aspect=1.5)

# Aadir ttulo y etiquetas
plt.title('Relacin entre Total de la Cuenta y Propina Separado por Sexo')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

# Mostrar el grfico
plt.show()
```

```python
# Crear un lmplot avanzado de la propina en funcin del total de la cuenta, separado por da de la semana y ajuste robusto
sns.lmplot(x='total_bill', y='tip', hue='sex', col='day', data=tips,
           height=4, aspect=0.8, robust=True, palette='coolwarm',
           markers=['o', 'x'], scatter_kws={'s': 30, 'alpha': 0.7})

# Aadir ttulo y etiquetas
plt.subplots_adjust(top=0.9)
plt.suptitle('Relacin entre Total de la Cuenta y Propina Separado por Sexo y Da de la Semana')

# Mostrar el grfico
plt.show()
```

```python
mpg = sns.load_dataset('mpg')
mpg
```

```python
#El parametro order= genera una recta de regresion de orden 2 (cuadratica)
print("El parametro order= genera una recta de regresion de orden 2 (cuadratica)")
sns.regplot(data=mpg, x='weight', y='mpg', order=2)
plt.show()

#El parametro lowess= suaviza aun mas la recta de regresion
# Utiliza una regresion no parametrica
print("El parametro lowess= suaviza aun mas la curva de regresion")

sns.regplot(data=mpg, x='horsepower', y='mpg', lowess=True)
plt.show()
```

```python
# Regresion logistica
# Este tipo de regresion se utiliza cuando la variable es de respuesta binaria

#La siguiente linea de codigo se explica asi:
#   en y= primero se utiliza la columna origin
#   luego eq() busca todos los resultados que digan 'usa'
#   por ultimo esta columna con resultados 'usa' se renombra como 'from_usa'


sns.regplot(x=mpg['weight'], y=mpg['origin'].eq('usa').rename("from_usa"), logistic=True)
plt.title("Regresion logistica")
plt.show()
```

```python
# Otro regplot es una regresion robusta
# que reduce la influencia de los valores atipicos

#Ejemplo regresion lineal simple
sns.regplot(data=mpg, x='horsepower', y='weight')
plt.title("Regresion lineal simple")
plt.show()

#Regresion lineal robusta

sns.regplot(data=mpg, x='horsepower', y='weight', robust=True)
plt.title("Regresion lineal robusta")
plt.show()

#Para este caso la diferencia es casi imperceptible
# pero si se observa los ultimos puntos donde cruza la
# recta de la regresion robusta se aprecia la minima diferencia
```

```python
# Regplot sin intervalo de confianza
# el intervalo de confianza es el sombreado que se aprecia
# alrededor de la recta en los graficos anteriores
sns.regplot(data=mpg, x='horsepower', y='weight', ci=None)
plt.title('Regplot sin intervalo de confianza')
plt.show()

#Esto es util para analizar rapido cuando hay grandes conjuntos de datos
```

```python
# Cuando la variable x sea discreta, se puede utilizar x_jitter=
#  para agregar algo de ruido uniforme para mejorar la visualizacion de datos discretos

sns.regplot(data=mpg, x='cylinders', y='weight', x_jitter=.15)
plt.title("Grafico con variable discreta y ruido agregado")
plt.show()
```

```python
import numpy as np
# Cargamos nuevamente los datos
mpg = sns.load_dataset('mpg')
```

```python
# Para agregacion cuando se tiene una
# x variable continua, se agrupa y luego agrega

sns.regplot(data=mpg, x='weight', y='mpg', x_bins= np.arange(2000,5500,250), order=2)
plt.title('Agrupacion y agregacion con linea de ajuste')
plt.show()
```

```python
# Tambien se puede tener mayor personalizacion

sns.regplot(data=mpg, x='weight', y='horsepower', ci=99, marker='x',
            color='.3', line_kws=dict(color='r'))
plt.title('Personalizacion del grafico y recta de regrecion')
plt.show()
```
