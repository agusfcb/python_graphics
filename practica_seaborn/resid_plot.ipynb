```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

```python
fig, axs = plt.subplots(2,2, figsize=(8,6))

sns.regplot(data=mpg, x='weight', y='displacement', ax=axs[0,0])
axs[0,0].set_title("Grafico de regresion de weight x displacement")

sns.residplot(data=mpg, x='weight', y='displacement', ax=axs[0,1])
axs[0,1].set_title("Grafico de residuos de weight x displacement")

#Ejemplo contrastando el grafico de regresion con el grafico de residuos de la misma regresion
sns.regplot(data=mpg, x='horsepower', y='mpg', ax= axs[1,0])
axs[1,0].set_title("Grafico de regresion de mpg x horsepower")

sns.residplot(data=mpg, x='horsepower', y='mpg', ax= axs[1,1])
axs[1,1].set_title("Grafico de residuos de mpg x horsepower")

plt.tight_layout()
plt.show()
```

```python
sns.residplot(data=mpg, x='horsepower', y='mpg')
plt.title("Residuo de horsepower x mpg")
plt.show()

# Suavisamos la curva con lowess y le damos el color rojo
sns.residplot(data=mpg, x='horsepower', y='mpg',  lowess=True, line_kws=dict(color='r'))
plt.title("Residuo de horsepower x mpg con linea de tendencia")
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
# Crear un residplot de los residuos de la propina en funcin del total de la cuenta
plt.figure(figsize=(8, 6))
sns.residplot(x='total_bill', y='tip', data=tips)

# Aadir ttulo y etiquetas
plt.title('Residuos de la Regresin entre Total de la Cuenta y Propina')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Residuos de la Propina')

# Mostrar el grfico
plt.show()

#Contrastamos contra el resdiplot con linea de tendencia
# el parametro lowess= suavisa los residuos
sns.residplot(x='total_bill', y='tip', data=tips, lowess=True, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red', 'lw': 2})

# Aadir ttulo y etiquetas
plt.title('Residuos de la Regresin entre Total de la Cuenta y Propina (Suavizado Lowess)')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Residuos de la Propina')

# Mostrar el grfico
plt.show()
```
