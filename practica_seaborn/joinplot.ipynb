```python
# Joinsplots como su nombre lo indica son la union de dos graficos

import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips' de Seaborn
tips = sns.load_dataset('tips')
```

```python
# Crear un jointplot
# Generalmente siempre tenemos dos componentes
#  El grafico interior que puede tambien contener varios graficos a la vez
#  Y el grafico exterior conocido como grafico marginal

sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', marginal_kws=dict(bins=15, fill=True))

# Mostrar el gr√°fico
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg', marginal_kws=dict(bins=20, kde=True))

plt.show()
```
