```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Establecer el tema
sns.set_theme(style='darkgrid')
# otros temas: white, dark, stick, whitegrid
```

```python
# Crear un grfico de dispersin
data = sns.load_dataset("tips")
sns.scatterplot(data=data, x="total_bill", y="tip", hue="day")
plt.show()
```

```python
#Lista de paletas de colores
palettes = ["deep", "muted", "bright", "pastel", "dark", "colorblind",
            "Blues", "BuGn", "BuPu", "GnBu", "Greens", "Greys", "Oranges", "OrRd", "PuBu", "PuBuGn", "PuRd", "Purples", "RdPu", "Reds", "YlGn", "YlGnBu", "YlOrBr", "YlOrRd",
            "afmhot", "autumn", "binary", "bone", "cool", "copper", "cubehelix", "flag", "gist_earth", "gist_gray", "gist_heat", "gist_ncar", "gist_rainbow", "gist_stern", "gist_yarg",
            "gnuplot", "gnuplot2", "gray", "hot", "hsv", "inferno", "magma", "nipy_spectral", "ocean", "pink", "plasma", "prism", "rainbow", "seismic", "spring", "summer", "tab10", "tab20", "tab20b", "tab20c", "terrain", "viridis", "winter"]

#Para probar paletas se utiliza la funcion color_palette

var_cubehelix = sns.color_palette('cubehelix')
sns.palplot(var_cubehelix)
var_viridis = sns.color_palette('viridis')
sns.palplot(var_viridis)
var_muted = sns.color_palette('muted')
sns.palplot(var_muted)

plt.show()
```

```python
# Graficando todas las paletas de colores disponibles en sns
n = len(palettes)
fig, axes = plt.subplots(ncols=1, nrows=n, figsize=(8, 2*n), dpi=80)

for ax, palette in zip(axes, palettes):
    colors = sns.color_palette(palette)
    n_colors = len(colors)
    ax.imshow([colors], aspect='auto')
    ax.set_xticks(range(n_colors))
    ax.set_xticklabels(range(1, n_colors+1))
    ax.set_yticks([])
    ax.set_title(palette, fontsize=12, pad=5)

plt.tight_layout()
plt.show()
```

```python
data = sns.load_dataset("iris")
sns.set_theme(style="whitegrid", palette="Blues")
sns.boxplot(data=data, x="species", y="sepal_length")
plt.show()
```

```python
sns.set_theme(style="dark", palette="pastel")

data = sns.load_dataset("penguins")
sns.violinplot(data=data, x="species", y="bill_length_mm", hue="sex", split=True)
plt.show()
```

```python
#Se puede personalizar el estilo y la paleta a la vez
sns.set_theme(style="ticks", palette="bright")

data = sns.load_dataset("flights")
sns.lineplot(data=data, x="year", y="passengers", hue="month")
plt.show()
```

```python
#Tambien se puede ajustar el contexto
# el contexto es la escala por defecto de todos los elementos del grafico
sns.set_context("paper")

# Crear un grfico de lnea
flights = sns.load_dataset("flights")
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.show()
```

```python
# Tambien se puede generar una paleta personalizada
custom_palette = sns.color_palette(["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"])
sns.set_theme(palette=custom_palette)

data = sns.load_dataset("diamonds")
sns.scatterplot(data=data, x="carat", y="price", hue="cut", marker='D')
plt.show()
```

```python
#Tambien se puede colocar un estilo de cuadricula
# y fondo con imagen
import numpy as np
from matplotlib.colors import ListedColormap

sns.set_theme(style="whitegrid")
data = sns.load_dataset("titanic")

cmap = ListedColormap(sns.color_palette("coolwarm", 256).as_hex())
sns.kdeplot(data=data, x="age", y="fare", cmap=cmap, fill=True)
plt.show()
```

```python
sns.set_theme(style="ticks", palette="cubehelix")

data = sns.load_dataset("tips")
sns.histplot(data=data, x="total_bill", hue="sex", element="step")
plt.show()
```

```python
sns.set_theme(style="white", palette="Set2")

data = sns.load_dataset("mpg")
sns.regplot(data=data, x="horsepower", y="mpg", scatter_kws={"s": 50}, line_kws={"color": "red"})
plt.show()
```

```python
sns.set_theme(style="dark", palette="colorblind", context="talk")
plt.figure(figsize=(12,8))

data = sns.load_dataset("planets")

sns.histplot(data=data, x="year", hue="method", multiple="stack")
plt.show()
```
