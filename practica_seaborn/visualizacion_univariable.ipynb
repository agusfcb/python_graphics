```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# Configurar el estilo de los grÃ¡ficos
sns.set(style="whitegrid")
# Crear un distplot para la variable 'total_bill'

plt.figure(figsize=(10,6))
sns.histplot(tips['total_bill'], kde=True, bins=20, color='skyblue')

# AÃ±adir tÃ­tulo y etiquetas
plt.title('DistribuciÃ³n del Total de la Cuenta con KDE')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Frecuencia')

# Mostrar el grÃ¡fico
plt.show()



```

```python
# personalizacion de kde con los datos ya cargados
import scipy as sp
```

```python
#Este es el primer grafico para usarlo de comparativa de algunos atributos del kde
sns.histplot(tips['total_bill'], kde=True, bins=20, color='skyblue')
plt.show()

# Originalmente el parametro para el relleno era shade pero se reemplazo por fill en sns

# fill es el relleno o sombra interna
# color es el color de la recta
# bw_adjust es el ancho de banda, que controla la suavidad de la curva KDE.
#   un valor menor hace que la curva sea mÃ¡s ajustada a los datos, mientras que un valor mayor la hace mÃ¡s suave.
sns.kdeplot(tips['total_bill'], fill=True, color='skyblue', bw_adjust=0.5)
plt.title('EstimaciÃ³n de Densidad del NÃºcleo (KDE) del Total de la Cuenta')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Densidad')
plt.show()
```

```python
# Ejemplo 3: KDE Acumulativo
plt.figure(figsize=(10, 6))

#cumulative=True
sns.kdeplot(tips['total_bill'], fill=True, color='skyblue', cumulative=True)
plt.title('EstimaciÃ³n de Densidad Acumulativa del Total de la Cuenta')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Densidad Acumulada')
plt.show()
```

```python
# Ejemplo 4: analisis KDE Bivariado

plt.figure(figsize=(10, 6))
sns.kdeplot(x=tips['total_bill'], y=tips['tip'], cmap='Blues', fill=True)
plt.title('EstimaciÃ³n de Densidad Bivariada del Total de la Cuenta y Propinas')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Propina ($)')
plt.show()
```
