```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Ejemplo de DataFrame
data = {
    'Region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Mes': ['Enero', 'Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Febrero'],
    'Ventas': [100, 150, 120, 110, 130, 140, 160, 155]
}
```

```python
df = pd.DataFrame(data)

# Creación de un gráfico de barras usando Seaborn
sns.barplot(x='Region', y='Ventas', data=df, hue='Mes')
plt.title('Ventas por Región y Mes')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python

# Configuración de estilo y tamaño de la figura
sns.set(style='whitegrid')  # Estilo de fondo del gráfico

plt.figure(figsize=(10, 6))  # Tamaño de la figura (ancho, alto)

# Crear un gráfico de barras de promedio de propinas por día
sns.barplot(x='day', y='tip', data=tips, palette='Set2')  # 'day' y 'tip' son columnas en 'tips'

# Añadir título y etiquetas de los ejes
plt.title('Promedio de Propinas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Promedio de Propina')

# Mostrar el gráfico
plt.show()
```
