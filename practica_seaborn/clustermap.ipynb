```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el dataset 'iris' de Seaborn
iris = sns.load_dataset('iris')
```

```python
# Seleccionar solo las columnas numricas
numeric_columns = iris.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlacin
correlation_matrix = numeric_columns.corr()

# Crear el clustermap avanzado
sns.clustermap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5, linecolor='black', cbar_kws={'label': 'Correlacin'})

# Aadir ttulo
plt.title('Clustermap de la Matriz de Correlacin del Dataset Iris', y=1.05)

# Mostrar el grfico
plt.show()
```

```python

# Cargar el dataset 'flights' de Seaborn
flights = sns.load_dataset('flights')

# Crear una tabla pivote con los datos de vuelos
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Crear el clustermap avanzado con los datos de vuelos
sns.clustermap(flights_pivot, cmap="YlGnBu", linewidths=.5, linecolor='black', cbar_kws={'label': 'Nmero de Pasajeros'}, annot=True, fmt='d')

# Aadir ttulo
plt.title('Clustermap del Nmero de Pasajeros por Mes y Ao', y=1.05)

# Mostrar el grfico
plt.show()
```

#### Explicacin del Cdigo
##### Carga y Preprocesamiento de Datos:

* En ambos ejemplos, se carga un dataset (iris y flights) usando sns.load_dataset.
* Para el dataset iris, se seleccionan solo las columnas numricas y se calcula la matriz de correlacin.
* Para el dataset flights, se crea una tabla pivote con los meses como filas, los aos como columnas y los nmeros de pasajeros como valores.


Creacin del Clustermap:

* sns.clustermap se utiliza para crear un mapa de calor con clustering jerrquico.
* cmap define la paleta de colores.
* linewidths y linecolor aaden lneas entre las celdas del heatmap.
* cbar_kws se usa para personalizar la barra de color.
* annot y fmt se utilizan para aadir anotaciones a las celdas del heatmap.


##### Personalizacin y Visualizacin:

* Se aaden ttulos a los grficos utilizando plt.title.
* y=1.05 en plt.title ajusta la posicin del ttulo para que no se superponga con el grfico.

## Diferencias entre heatmap y clustermap
La diferencia fundamental es el orden.

Mientras que un heatmap respeta estrictamente el orden de tus datos (tal como vienen en tu DataFrame), un clustermap los reordena y agrupa automticamente basndose en qu tan similares son las filas y las columnas entre s.

* El orden de un heatmap: Si tu DataFrame tiene la fila "Enero" primero y "Diciembre" al final, el grfico las mostrar en ese orden exacto.
* El orden de un clustermap: Destruye tu orden original. Si "Enero" se parece ms a "Marzo" que a "Febrero", los pondr juntos.
