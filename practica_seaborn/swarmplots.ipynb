```python
# Se utilizan para ver distribuciones de graficos sin superposiciones de sus puntos
# Son buen complemento junto con otro graficos
# Funcionan bien en combinacion con otros tipos de graficos como boxplots

import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips' de Seaborn
tips = sns.load_dataset('tips')

# Crear un swarmplot de la propina por día de la semana
plt.figure(figsize=(8, 6))
sns.swarmplot(x='day', y='tip', data=tips)

# Añadir título y etiquetas
plt.title('Distribución de Propinas por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Propina')

# Mostrar el gráfico
plt.show()
```

```python
# Crear un swarmplot avanzado de la propina por día de la semana, separado por sexo
plt.figure(figsize=(10, 6))

# Añadir un violinplot de fondo
# inner= elimina los elementos internos del violinplot como el boxplot y los bitoges
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray')

# Añadir el swarmplot
# hue= agrupa por una variable categorica del dataset
# dodge=True hace que los puntos agrupados por hue= tambien se vean visiblemente separados en el grafico
# palette= se utiliza para personalizar con una paleta de colores que trae sns
sns.swarmplot(x='day', y='tip', hue='sex', data=tips, palette='Set1', dodge=True)

# Añadir título y etiquetas
plt.title('Distribución de Propinas por Día de la Semana Separado por Sexo')
plt.xlabel('Día de la Semana')
plt.ylabel('Propina')

# Eliminar duplicados de leyenda
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[:2], labels[:2])

# Mostrar el gráfico
plt.show()
```

```python
#Otro ejemplo
# Distribucion del total de la mesa por la cantidad de personas
plt.figure(figsize=(10, 6))

sns.swarmplot(x='total_bill', y='day', data=tips, hue='size')

plt.xlabel('total_bill')
plt.ylabel('day')
plt.title("Distribucion del total de la mesa por la cantidad de personas")
plt.show()
```

```python
#Otro ejemplo
# Distribucion del total de la mesa por la cantidad de personas
plt.figure(figsize=(10, 6))

sns.swarmplot(x='total_bill', y='day', data=tips, hue='size', palette='crest', dodge=True)

plt.xlabel('total_bill')
plt.ylabel('day')
plt.title("Distribucion del total de la mesa por la cantidad de personas")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.swarmplot(x='total_bill', y='day', data=tips, hue='size', palette='crest', dodge=True)

plt.xlabel('total_bill')
plt.ylabel('day')
plt.title("Distribucion del total de la mesa por la cantidad de personas")
plt.show()
```

```python
# Tambien se pueden realizar graficos mas enfocados en un grupo de datos
# Esto se hace con la funcion query()

plt.figure(figsize=(10, 6))
#La siguiente linea muestra solo los datos donde size sea 2, 3 y 5
sns.swarmplot(data= tips.query("size in [2, 3, 5]"), x='total_bill', y='size', orient='h')
```

```python
# Un swarmplot tambien puede ser argumento de otros graficos como catplots

plt.figure(figsize=(10, 6))

#Aqui el argumento es kind='swarm'
sns.catplot(data=tips, kind='swarm', x='time', y='total_bill', hue='sex', col='day', aspect=.5)
plt.show()
```
