### Colores en Matplotlib
Uso de Colores Predefinidos

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Se puede personalizar usando colores predefinidos con los nombres en ingles
# Pero tambien se puede personalizar utilizando las especificaciones de colores HEX

fig, ax = plt.subplots(1,2, figsize=(10, 5))

ax[0].plot(x, y, color='blue')  # Uso de un color predefinido
plt.title('Uso de Colores Predefinidos')
ax[1].plot(x, y, color='green')


plt.show()
```

Especificación de Colores con Nombres, Códigos HEX y RGB


```python
plt.plot(x, y, color='#FF5733')  # Usando un código HEX
plt.title('Especificación de Color con HEX')
plt.show()

plt.plot(x, y, color=(0.1, 0.2, 0.5))  # Usando valores RGB
plt.title('Especificación de Color con RGB')
plt.show()
```

### Mapa de Colores (Colormaps)
Los colormaps son una forma de mapear valores numéricos a colores.

```python
import numpy as np

z = np.linspace(0, 1, 100)
colores = plt.cm.viridis(z)

for i in range(len(z)):
    plt.plot(x, [v * (i+1) for v in y], color=colores[i])

plt.title('Uso de Colormaps')
plt.show()
```

### Estilos de Línea
Tipos de Líneas Disponibles


```python
plt.plot(x, y, linestyle='-')  # Línea sólida
plt.plot(x, y, linestyle='--')  # Línea discontinua
plt.plot(x, y, linestyle='-.')  # Línea punto y raya
plt.plot(x, y, linestyle=':')  # Línea punteada
plt.title('Tipos de Líneas Disponibles')
plt.show()
```

### Marcadores
Tipos de Marcadores Disponibles

```python
plt.plot(x, y, marker='o')  # Círculo
plt.plot(x, y, marker='s')  # Cuadrado
plt.plot(x, y, marker='^')  # Triángulo
plt.title('Tipos de Marcadores Disponibles')
plt.show()
```

Combinación de Marcadores y Estilos de Línea


```python
plt.plot(x, y, linestyle='--', marker='o', color='purple')
plt.title('Combinación de Marcadores y Estilos de Línea')
plt.show()
```

### Técnicas Avanzadas de Personalización
Ancho de Línea (linewidth)

```python
plt.plot(x, y, linewidth=3, color='green')
plt.title('Ancho de Línea')
plt.show()
```

Transparencia (alpha)


```python
plt.plot(x, y, color='red', alpha=0.5)
plt.title('Transparencia de Línea')
plt.show()
```

Uso de Estilos de Línea Personalizados


```python
linea_personalizada = (0, (5, 10))  # 5 píxeles dibujados, 10 píxeles en blanco
plt.plot(x, y, linestyle=linea_personalizada, color='black')
plt.title('Uso de Estilos de Línea Personalizados')
plt.show()
```

### Ejemplos Prácticos
Ejemplo Básico


```python
plt.plot(x, y, color='blue', linestyle='-', marker='o')
plt.title('Ejemplo Básico de Personalización')
plt.show()
```

Ejemplo Avanzado con Personalización Detallada


```python
plt.plot(x, y, color='#FF5733', linestyle='-.', linewidth=2, marker='^', markersize=10, alpha=0.7)
plt.title('Ejemplo Avanzado con Personalización Detallada')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```
