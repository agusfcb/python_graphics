```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#crea el contenedor principal
fig = plt.figure(figsize=(10, 6))

# agrega un grafico (plot) al contenedor
ax = fig.add_subplot(1, 1, 1)

#edita el grafico
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

line = ax.plot(x, y, label='Seno')
text = ax.text(2, np.sin(2), 'Punto', fontsize=12, color='blue')
ax.legend()
plt.show()
```

```python
# Nuevos datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
```

```python
#estilos disponibles que traen predeterminados
#para elegir matplotlib y seaborn
print("Estilos disponibles:", plt.style.available)

#seleccion del estilo
plt.style.use('seaborn-v0_8-notebook')

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, label='Seno', color='blue')
ax.plot(x, y2, label='Coseno', color='red')
# Opcional: ax.grid(True, linestyle='--', color='grey', alpha=0.7)
ax.set_title('Grfico de Seno y Coseno con estilo seaborn-darkgrid')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
plt.show()
```

```python
# Probando otro estilo
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, label='Seno', color='green')
ax.plot(x, y2, label='Coseno', color='orange')
ax.set_facecolor('lightgrey')
ax.grid(True, linestyle='--', color='grey', alpha=0.7)
ax.set_title('Grfico de Seno y Coseno con estilo ggplot modificado')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()
plt.show()
```
