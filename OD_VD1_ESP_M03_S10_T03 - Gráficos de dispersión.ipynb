### Creación de Gráficos de Dispersión Básicos

Importación de Bibliotecas Necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de Datos de Ejemplo

```python
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + 1 + np.random.normal(0, 0.1, 50)
```

Creación de un Gráfico de Dispersión Básico


```python
plt.scatter(x, y, color='blue')
plt.title('Gráfico de Dispersión Básico')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

### Representación de Variables Correlacionadas
Cálculo y Visualización de la Correlación



```python
correlacion = np.corrcoef(x, y)[0, 1]
plt.scatter(x, y, color='blue')
plt.title(f'Gráfico de Dispersión (Correlación: {correlacion:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

Añadir Líneas de Tendencia


```python
plt.scatter(x, y, color='blue', label='Datos')
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label='Línea de Tendencia')
plt.title(f'Gráfico de Dispersión con Línea de Tendencia (Correlación: {correlacion:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```

### Personalización de Gráficos de Dispersión
Cambiar Colores y Estilos de Puntos


```python
plt.scatter(x, y, color='purple', marker='^')
plt.title('Gráfico de Dispersión Personalizado')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

Añadir Títulos y Etiquetas


```python
plt.scatter(x, y, color='green')
plt.title('Gráfico de Dispersión con Títulos y Etiquetas')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

Personalización de Ejes y Leyendas


```python
plt.scatter(x, y, color='orange', label='Datos')
plt.title('Gráfico de Dispersión con Leyenda')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.grid(True)
plt.show()
```

### Técnicas Avanzadas en Gráficos de Dispersión
Uso de Tamaños y Colores para Representar Información Adicional


```python
sizes = 1000 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.title('Gráfico de Dispersión Avanzado')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.colorbar()  # Añadir barra de color para los colores
plt.show()
```

Creación de Gráficos de Dispersión 3D


```python
from mpl_toolkits.mplot3d import Axes3D

z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='blue', marker='o')

ax.set_title('Gráfico de Dispersión 3D')
ax.set_xlabel('Variable X')
ax.set_ylabel('Variable Y')
ax.set_zlabel('Variable Z')

plt.show()
```

Subplots con Gráficos de Dispersión


```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].scatter(x, y, color='red')
axs[0].set_title('Gráfico de Dispersión 1')
axs[0].set_xlabel('Variable X')
axs[0].set_ylabel('Variable Y')

axs[1].scatter(y, x, color='blue')
axs[1].set_title('Gráfico de Dispersión 2')
axs[1].set_xlabel('Variable Y')
axs[1].set_ylabel('Variable X')

plt.tight_layout()
plt.show()
```
