```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#datos
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
from matplotlib.pyplot import xlabel

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x,y, label="Seno(x)",color='blue')

ax.set_title('Grafico del seno')
ax.set_xlabel('Longitud')
ax.set_ylabel('Amplitud')
plt.grid()
plt.legend()

plt.show()
```

```python
#Tambien se pueden utilizar metodos universales de pyplot sin
# referir a un objeto "ax", pero el objeto ya tiene que estar creado
# y pyplot aplicara los cambios al ultimo objeto activo (si existen varios graficos)
from matplotlib.lines import lineStyles

plt.plot(x, y, label="Seno(x)", linestyle='--', color='red')
plt.title("Nuevo grafico del coseno")
plt.xlabel("Longitud")
plt.ylabel("Amplitud")
plt.grid()
plt.legend()
plt.show()
```

```python
#Tambien se puede graficar multiples lineas en un mismo grafico
y2 = np.cos(x)
plt.plot(x,y, label="Seno(x)", color='red')
plt.plot(x, y2, label="Coseno(x)", color='blue')
plt.title("Grafico de multiples lines: seno y coseno")
plt.legend()
plt.show()
```

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)

# 1. Definimos la estructura: 2 filas, 2 columnas
# figsize es clave para que no se vean apretados
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Subplots devolvio una matriz de 2x2 que se guarda en axs
# 2. APLANAMOS la matriz de ejes para iterar fcil
# esto se hace con la funcion flatten que transforma una matriz en una lista
# Ahora 'lista_axs' es [ax1, ax2, ax3, ax4] en lugar de una matriz 2x2
lista_axs = axs.flatten()

# Configuraciones diferentes para cada grfico
estilos = ['-', '--', '-.', ':']
colores = ['blue', 'orange', 'green', 'red']
titulos = ['Lineal', 'Cuadrtica', 'Cbica', 'Logartmica']

# 3. Iteramos y dibujamos
for i, ax in enumerate(lista_axs):
    y = x ** (i + 1) # esta formula genera diferentes curvas: lineal, cuadratica, etc.
    
    # Dibujamos en el eje especfico 'ax'
    ax.plot(x, y, color=colores[i], linestyle=estilos[i])
    
    # Configuracin independiente
    ax.set_title(titulos[i])
    ax.set_xlabel('Eje X comn')
    ax.grid(True, alpha=0.3)

# 4. Ajuste final (Muy importante!)
# Evita que los ttulos se monten sobre los ejes de arriba
plt.tight_layout()
plt.show()
```

```python
#Otro ejemplo con subplot_mosaic

# Definimos el diseo visualmente con una lista de listas
# "A A": La grfica A ocupa todo el ancho arriba
# "B C": Abajo se dividen B y C
layout = [
    ["A", "A"],
    ["B", "C"]
]

fig, axs_dict = plt.subplot_mosaic(layout, figsize=(10, 6))

# Ahora 'axs_dict' es un diccionario: {'A': ax_obj, 'B': ax_obj, ...}

# Grfico A (El grande)
axs_dict['A'].plot([1, 2, 3], [1, 4, 9], 'tab:blue')
axs_dict['A'].set_title("Visin General (Grfico A)")

# Grfico B (Abajo izquierda)
axs_dict['B'].scatter([1, 2, 3], [3, 2, 1], color='tab:red')
axs_dict['B'].set_title("Detalle 1 (Grfico B)")

# Grfico C (Abajo derecha)
axs_dict['C'].bar(['x', 'y', 'z'], [10, 20, 15], color='tab:green')
axs_dict['C'].set_title("Detalle 2 (Grfico C)")

plt.tight_layout()
plt.show()
```

```python
#nuevo grafico
#datos
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, axs = plt.subplots(figsize=(10, 6))

axs = plt.plot(x, y)
axs = plt.title('Grfico con Anotaciones')
axs = plt.xlabel('Eje X')
axs = plt.ylabel('Eje Y')

# La siguiente linea agrega una anotacion con una flecha.
#    xy= recibe como argumento la ubicacion de x e y del punto a anotar
#    xytext= recibe como argumento la ubicacion del texto, puede estar fuera de los puntos visibles del grafico
#    arrowprops= recibe las propiedades de la flecha que apunta al punto anotado
#    shrink=0.05 indica que la flecha se acorta un 5% para no tocar el punto exacto
axs = plt.annotate('Punto mximo', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
axs = plt.show()
```
