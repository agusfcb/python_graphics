```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
plt.boxplot(datos)
plt.title('Boxplot BÃ¡sico')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

```python
# Tambien se puede personalizar de forma mas compleja
plt.boxplot(datos, patch_artist=True, labels=['A', 'B', 'C'], #patch_artist=True es el interruptor que permite rellenar de color las cajas
            boxprops=dict(facecolor='lightgreen', color='green'), # facecolor es el interior de la caja y color es el borde
            whiskerprops=dict(color='green'), # Bigotes
            capprops=dict(color='black'), # Limites de cuartiles
            medianprops=dict(color='red')) # Mediana
plt.title('Boxplot con Leyenda y Etiquetas')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

```python
# Otro ejemplo mas complejo

# Generador de datos
datos_agrupados = [np.random.normal(loc, scale, 100) for loc, scale in zip(range(1, 4), range(1, 4))]

plt.boxplot(datos_agrupados, patch_artist=True, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'],
            boxprops=dict(facecolor='lightcoral', color='darkred'),
            whiskerprops=dict(color='darkred'),
            capprops=dict(color='darkred'),
            medianprops=dict(color='yellow'),
            vert=False)
plt.title('Boxplot con Datos Agrupados')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

```python
#Un tipo de grafico similar al boxplot es el violinplot
# este muestra un poco mas de informacion sobre la distribucion de los datos
# se puede observar la densidad de los datos a lo largo del rango de valores
plt.violinplot(datos_agrupados, showmeans=True)
plt.title('GrÃ¡fico de ViolÃ­n con Datos Agrupados')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.show()
```

```python
# Tambien se pueden presentar varios estadisticos como por ejemplo la media y la mediana
# aunque para el ejemplo ambos valores son muy cercanos por eso no se notan por separado
plt.violinplot(datos, showmeans=True, showmedians=True)
plt.title('Ejemplo Avanzado de GrÃ¡fico de ViolÃ­n')
plt.xlabel('Grupo')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```
