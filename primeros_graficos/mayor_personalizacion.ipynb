```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Se puede personalizar usando colores predefinidos con los nombres en ingles
# Pero tambien se puede personalizar utilizando las especificaciones
#  de colores HEX y RGB

fig, ax = plt.subplots(2,2, figsize=(10, 5))

ax[0, 0].plot(x, y, color='green')  # Uso de un color predefinido
ax[0, 0].set_title('Uso de Colores Predefinidos')

ax[0, 1].plot(x, y, color='#FF5733') #Uso de color con HEX
ax[0, 1].set_title('Uso de Colores HEX')

ax[1, 0].plot(x, y, color=(0.1, 0.2, 0.5)) #Uso de color con RGB
ax[1, 0].set_title('Uso de Colores RGB')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

z = np.linspace(0, 1, 100)
colores = plt.cm.viridis(z)

for i in range(len(z)):
    plt.plot(x, [v * (i+1) for v in y], color=colores[i])
plt.title('Uso de Colormaps')
plt.show()
```

```python
# Existen muchos mapas de colores
# Se pueden observar las opciones escribiendo cm.
z = np.linspace(0, 1, 100)
colores = plt.cm.Blues(z)

for i in range(len(z)):
    plt.plot(x, [v * (i+1) for v in y], color=colores[i])

plt.title('Uso de Colormaps')
plt.show()
```

```python
# Para personalizar las lineas existen opciones como las siguientes

#plt.plot(x, y, linestyle='-')  # Lnea slida
#plt.plot(x, y, linestyle='--')  # Lnea discontinua
plt.plot(x, y, linestyle='-.')  # Lnea punto y raya
#plt.plot(x, y, linestyle=':')  # Lnea punteada
plt.title('Tipos de Lneas Disponibles')
plt.show()
```

```python
#PAra personalizar los marcadores de los puntos
plt.plot(x, y, marker='o')  # Crculo
#plt.plot(x, y, marker='s')  # Cuadrado
#plt.plot(x, y, marker='^')  # Tringulo
plt.title('Tipos de Marcadores Disponibles')
plt.show()
```

```python
#Ejemplo combinando lineas, marcadores y otros atributos
fig, axs = plt.subplots(1, 2, figsize=(8, 5))


axs[0].plot(x, y, linestyle='--', marker='o', linewidth=2, color='purple')

axs[0].set_title('Combinacin de Marcadores y Estilos de Lnea')

axs[1].plot(x, y, color='red', alpha=0.5)
axs[1].set_title('Transparencia de Lnea')

plt.tight_layout(w_pad=5)
plt.show()
```

```python
#Ejemplo basico de personalizacion
linea_personalizada = (0, (5, 10))  # 5 pxeles dibujados, 10 pxeles en blanco
plt.plot(x, y, linestyle=linea_personalizada, color='black')
plt.title('Uso de Estilos de Lnea Personalizados')
plt.show()
```

```python
#Ejemplo avanzado de personalizacion
plt.plot(x, y, color='#FF5733', linestyle='-.', linewidth=2, marker='^', markersize=10, alpha=0.7)
plt.title('Ejemplo Avanzado con Personalizacin Detallada')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```
