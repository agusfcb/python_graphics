```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + 1 + np.random.normal(0, 0.1, 50)
```

```python
#Scatter es la funcion para graficos de dispersion
plt.scatter(x, y, color='blue')
plt.title('Gráfico de Dispersión Básico')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
correlacion = np.corrcoef(x, y)[0, 1]
#Se agrega la correlación al título del gráfico
plt.scatter(x, y, color='blue')
plt.title(f'Gráfico de Dispersión (Correlación: {correlacion:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
#Primero se dibujan los puntos
plt.scatter(x, y, color='blue', label='Datos')

#Luego se calcula la linea de tendencia
# polyfit devuelve los coeficientes de la linea de mejor ajuste
# 1 es el grado del polinomio, en este caso grado 1 es un polinomio lineal, 2 seria cuadratico, etc.
# con la formula general de rectas la funcion devuelve m (pendiente) y b(interseccion)
m, b = np.polyfit(x, y, 1)

#Luego graficamos la linea sobre el grafico de dispersion ya existente
plt.plot(x, m*x + b, color='red', label='Línea de Tendencia')

plt.title(f'Gráfico de Dispersión con Línea de Tendencia (Correlación: {correlacion:.2f})')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.legend()
plt.show()
```

```python
# Se puede personalizar otros estilos de puntos en el grafico de dispersion con marker
plt.scatter(x, y, color='purple', marker='^')
plt.title('Gráfico de Dispersión Personalizado')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
#Una grafica interesante es la del grafico de dispersion con tamaños y colores variables

#Creamos tamaños variables
sizes = 1000 * np.random.rand(50)

#Creamos valores de colores variables, esto se usara como argumento para el parametro color
# como se observa el argumento c= admite una lista de enteros que estan asociados a colores
# la diferencia entre color= y c= es que color= define un unico color para todos los puntos y c= puede utilizarse para definir varios colores en el mismo grafico
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.title('Gráfico de Dispersión Avanzado')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.colorbar()  # Añadir barra de color para los colores
plt.show()
```

```python
#Para graficar en 3D hay que importar la libreria especial mplot3d
from mpl_toolkits.mplot3d import Axes3D

z = np.random.rand(50)

fig = plt.figure()
#Aqui se le dice a subplot que es un grafico de proyeccion 3d
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='blue', marker='o')

ax.set_title('Gráfico de Dispersión 3D')
ax.set_xlabel('Variable X')
ax.set_ylabel('Variable Y')
ax.set_zlabel('Variable Z')

plt.show()
```

```python
#Como ya se vio se pueden crear multiples graficos en una solo espacio/figura
#El argumento 2 es el numero de columnas
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].scatter(x, y, color='red')
axs[0].set_title('Gráfico de Dispersión 1')
axs[0].set_xlabel('Variable X')
axs[0].set_ylabel('Variable Y')

axs[1].scatter(y, x, color='blue')
axs[1].set_title('Gráfico de Dispersión 2')
axs[1].set_xlabel('Variable Y')
axs[1].set_ylabel('Variable X')

plt.tight_layout()
plt.show()
```

```python
#Por ultimo tambien se puede personalizar el grafico de dispersion con tamaño con algun estilo
# En este caso el mapa de colores a utilizar se define con cmap=

sizes = 1000 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='plasma')
plt.title('Ejemplo Avanzado de Gráfico de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.colorbar(label='Intensidad')
plt.grid(True)
plt.show()
```
