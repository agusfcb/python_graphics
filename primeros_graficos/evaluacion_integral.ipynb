```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df
```

* carat: peso del diamante en quilates. (Numero)
* cut: calidad del corte del diamante. (5 categorias)
* color: color del diamante. (10 categorias)
* clarity: claridad del diamante. (8 categorias)
* depth: Profundidad total del diamante como un porcentaje del promedio del diámetro. (Numero)
* table: Tamaño de la mesa del diamante (parte superior plana) como un porcentaje del diámetro del promedio. (Numero)
* price: Precio del diamante en dólares estadounidenses. (Numero)
* x: Longitud del diamante en milímetros. (Numero)
* y: Ancho del diamante en milímetros. (Numero)
* z: Altura del diamante en milímetros. (Numero)

```python
#Categorias de precios
bins = [0, 500, 1000, 5000, 10000, 20000]
labels = ['Bajo', 'Medio-Bajo', 'Medio', 'Medio-Alto', 'Alto']
df['price_category'] = pd.cut(df['price'], bins=bins, labels=labels)
```

```python
sns.histplot(df['price'])
```

```python
sns.scatterplot(data=df, x='x', y='price')
```

```python

sns.boxplot(df, x='clarity', y='depth')
plt.show()
```

```python
sns.countplot(df, x='price_category')
```

```python
sns.scatterplot(data=df, x='table', y='depth')
```

```python
fig = plt.figure(figsize=(12,8))
sns.set_context('paper')
sns.violinplot(data=df, x='cut', y='price', hue='cut')
```

```python
sns.scatterplot(data=df, x='carat', y='depth', hue='color')
```

```python
sns.histplot(data=df, x='cut', hue='price_category', multiple='stack')
```
