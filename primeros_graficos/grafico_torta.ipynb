```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
etiquetas = ['A', 'B', 'C', 'D']
tamaños = [15, 30, 45, 10]
```

```python
plt.pie(tamaños, labels=etiquetas)
plt.title('Gráfico de Pie Básico')
plt.show()
```

```python
# Para mostrar procentajes se utiliza el parametro autopct
# el porcentaje es el dato de tamaño por lo que hay que tener en cuenta
# que debe sumar 100%
plt.pie(tamaños, labels=etiquetas, autopct='%1.1f%%')
plt.title('Gráfico de Pie con Porcentajes')
plt.show()
```

```python
# Al igual que otros graficos se puede personalizar
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90) # startangle es el angulo desde el que arranca
plt.title('Gráfico de Pie Personalizado')
plt.show()
```

```python
explosion = (0.1, 0, 0, 0.1)  # cada indice corresponde a un sector del grafico
# y el numero corresponde a la distancia que separa el sector
# en este ejemplo separo el amarillo en 0.1 y el celeste tambien en 0.1
plt.pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140, explode=explosion)
plt.title('Gráfico de Pie con Explosión de Sector')
plt.show()
```

```python
#Por ultimo colocando varios graficos en una figura
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Gráfico de Pie 1')

axs[1].pie(tamaños, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Gráfico de Pie 2')

plt.tight_layout()
plt.show()
```
